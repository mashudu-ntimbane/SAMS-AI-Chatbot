## Step 2 — Expose API with ngrok
ngrok creates a public URL for your local Flask server — useful for testing with PHP.

```python
from pyngrok import ngrok
import threading

# Run Flask in a background thread
# (import chatbot and app from 02_api_server.py first)

def run_flask():
    app.run(port=5000, debug=False, use_reloader=False)

t = threading.Thread(target=run_flask, daemon=True)
t.start()

# Create public URL (get free token at ngrok.com)
# ngrok.set_auth_token('YOUR_NGROK_TOKEN')
public_url = ngrok.connect(5000)
print(f'\n🌐 Public chatbot URL: {public_url}/chat')
print('Use this URL in your PHP CHATBOT_API_URL constant')
```

## Step 3 — Test the API

```python
import requests

test_questions = [
    'How do I pay my rent?',
    'Can I have overnight guests?',
    'Is my application approved?',
    'Are pets allowed?',
    'Hello there!'
]

for q in test_questions:
    res = requests.post('http://localhost:5000/chat', json={'message': q})
    data = res.json()
    print(f'Q: {q}')
    print(f'Intent: {data["intent"]} | Method: {data["method"]} | Confidence: {data["confidence"]}')
    print(f'A: {data["response"][:80]}...')
    print('---')
```
